```python
import pandas as pd
```

```python
taxi_zone = pd.read_csv('taxi+_zone_lookup.csv')
taxi_zone1 = taxi_zone.loc[taxi_zone.Borough.isin(["Brooklyn","Manhattan","Queens","Bronx","Staten Island"])]
taxi_zone1
```

```python
taxi_zone1_Astoria = taxi_zone1.loc[taxi_zone1.Zone == 'Astoria']
taxi_zone1_Astoria
```

```python
df1 = pd.read_csv('green_tripdata_2019-09.csv.gz', compression='gzip')
df1.head()
```

```python
## Question 3. Count records 

df1_filtered = df1.loc[(df1.lpep_pickup_datetime >= '2019-09-18 00:00:00') & (df1.lpep_dropoff_datetime <= '2019-09-18 23:59:59')]
df1_filtered
```

```python
## Question 4. Largest trip for each day

max_distance = df1['trip_distance'].max()
df_max_distance = df1[df1['trip_distance'] == max_distance]
df_max_distance
```

```python

## Question 5. The number of passengers
df1_filtered1 = df1.loc[(df1.lpep_pickup_datetime >= '2019-09-18 00:00:00') & (df1.lpep_pickup_datetime <= '2019-09-18 23:59:59')]
merged_df = pd.merge(df1_filtered1, taxi_zone1, left_on='PULocationID', right_on='LocationID')
merged_df
locations = ["Brooklyn", "Manhattan", "Queens", "Bronx", "Staten Island"]
filtered_df = merged_df[merged_df['Borough'].isin(locations)]

# Group by location and sum total_amount
result = filtered_df.groupby('Borough')['total_amount'].sum().reset_index()

# Display the result
result
```

```python
# ## Question 6. Largest tip

df1_filtered2 = df1.loc[(df1.lpep_pickup_datetime >= '2019-09-01 00:00:00') & (df1.lpep_pickup_datetime <= '2019-09-30 23:59:59')]
merged_df = pd.merge(df1_filtered2, taxi_zone1, left_on='PULocationID', right_on='LocationID')
merged_df
Zone_check = ["Astoria"]
filtered_df = merged_df[merged_df['Zone'].isin(Zone_check)]
max_tip = max(filtered_df.tip_amount)
largest_tip_dropofflocation = filtered_df.loc[filtered_df.tip_amount == max_tip].DOLocationID
taxi_zone.loc[taxi_zone.LocationID == largest_tip_dropofflocation.values[0]]
```
